제1유형
1. 이동시간(출동시간과 도착시간의 차이)이 평균적으로 가장 오래걸린 구급대의 이동시간을 구하시오.(분단위로 환산 및 반올림하여 정수로 출력)

```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_1.csv")
df
```

```python
df.info()
```

```python
# 데이터 타입 변경
df['출동시간']=pd.to_datetime(df['출동시간'])
df['도착시간']=pd.to_datetime(df['도착시간'])
```

```python
# 도착시간과 출동시간의 차이(분)
df['이동시간'] = df['도착시간'] - df['출동시간']
df['이동시간'] = df['이동시간'].dt.total_seconds()/60
print(df)
```

```python
# 구급대별 이동시간의 평균
group_mean = df.groupby('구급대')[['이동시간']].mean()
print(group_mean)
```

```python
# 이동시간이 가장 큰 값
result = round(group_mean['이동시간'].max())
print(result)
```

2. 주어진 데이터를 이용하여 교수 1인당 맡고 있는 학생 수가 가장 많은 대학교를 찾고 그 대학교의 전체 교수의 수를 구하시오(정수로 출력)

```python
# 데이터 불러오기
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_2.csv")
df
```

```python
# 전체학생수
df['학생']=df.iloc[:,2:].sum(axis=1)
df
```

```python
# 교수 1인당 맡은 학생수
df['교수당학생수']=df['학생']/df['교수']
df
```

```python
# 가장 큰 값을 가진 대학교 교슈의 수
max_id = df['교수당학생수'].idxmax()
result = int(df.loc[max_id,'교수'])
print(result)
```

3. 주어진데이터를 이용하여 연도별 총 범죄 건수의 월평균 값을 구하고 그 중 가장 큰 값 출력하기(반올림하여 정수로 출력)

```python
# 패키지 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
df=pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_3.csv")
df
```

```python
# 총합, 연도 항목 정의
df['총합']=df.iloc[:,1:7].sum(axis=1)
df['연도']=df['날짜'].str[:4]
```

```python
# 연도별 총합의 월평균 출력
year_mean=df.groupby('연도')['총합'].mean()
print((year_mean))
```

```python
# 가장 큰 값
result=round(year_mean.max())
print(result)
```

```python
# 패키지 임포트
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

# test 데이터 불러오기
test_df=pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_test.csv')
print(test_df)

# train 데이터 불러오기
train_df=pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_2_train.csv')
print(train_df)
```

```python
# object type 관련 데이터 확인
train_df.info()
train_df.describe(include='object')

# train의 결측치와 학생별 분포 확인
train_df.isnull().sum()
train_df["구분"].value_counts()

# test의 데이터 type과 결측치 확인
test_df.describe(include='object')
test_df.isnull().sum()

# 구분 열을 타켓 값으로 분리 및 구분 제거 데이터 출력
targetVal=train_df.pop('구분')
print(train_df.shape, test_df.shape)

# 인코딩 진행
train_df=pd.get_dummies(train_df)
test_df=pd.get_dummies(test_df)
print(train_df, test_df)
```

```python
# 검증용 데이터 분할
from sklearn.model_selection import train_test_split
train_x,validation_x,train_y,validation_y=train_test_split(train_df,targetVal,test_size=0.2,random_state=0)
print(train_x.shape,validation_x.shape,train_y.shape,validation_y.shape)

# 분류 모델 불러오기
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import  f1_score

# 분류 모델 생성 및 평가
model=RandomForestClassifier(random_state=0)
model.fit(train_x,train_y)
prediction_val=model.predict(validation_x)
print(f1_score(validation_y,prediction_val,average='macro'))

```

```python
# 테스트용 데이터를 이용해서 예측
prediction_test=model.predict(test_df)
result=pd.DataFrame(prediction_test)
result.to_csv('result.csv')
```
